## **Part 1: Prompt another person to create prompts for an AI. (9 pts)**

In this part, you will study how **guidance changes someone’s ability to write an effective AI prompt**.

Select a specific or moderately complex AI text-generation task—something that requires the model to make decisions about structure, tone, or reasoning rather than simple restatement. 

Some examples include:
- Summarize a research abstract into a tweet while keeping technical accuracy.
- Prompt an AI to explain how gradient descent works in simple terms, then in formal mathematical language.
- Write a 100-word story where the main character is a sentient vending machine that learns to feel emotions.
- Create a marketing tweet for a fake tech startup that sounds confident but subtly untrustworthy.

Then, ask a peer to write prompts for that task under two different conditions:

* **Condition 1:** They write their prompt freely in a **plain textbox** (no hints or structure).
* **Condition 2:** You give them guidance (written, visual, interactive, etc.) — for example, a short list of reminders, a color-coded example showing how to structure a clear prompt, or an interactive checklist that asks for task, audience, and tone. This should simulates what a guided interface *might* offer.

Use the two prompts your peer created to query an AI text-generation tool such as:

* [https://chat.openai.com](https://chat.openai.com)
* [https://claude.ai](https://claude.ai)
* [https://www.meta.ai](https://www.meta.ai)
* [https://deepai.org](https://deepai.org)
* [https://transformer.huggingface.co](https://transformer.huggingface.co)
