# Privacy on Canvas

---


In this assignment you will consider privacy-utility tradeoffs in the design of ed tech software like Canvas.

---


Canvas allows instructors to view “access reports” for each student. These access reports contain information about what content each student has viewed, which course files each student has downloaded, etc.
Screen Shot 2022-11-02 at 3.16.22 PM.png
Instructors can also view "quiz logs," which provide detailed information about students' behavior while taking quizzes.
Screen Shot 2022-11-03 at 7.30.04 AM.png

### 1: Do you think Canvas’ access reports and quiz logs provide an appropriate balance between utility for instructors and privacy for students? Why or why not? (4 pts)

### 2: Sketch two designs for a new feature that would replace Canvas access reports or quiz logs—still providing instructors with some information about student engagement but also providing better privacy for students. Each sketch should be accompanied with an explanation in 2-3 sentences of how the design balances student privacy against instructors' information needs. (8 pts per sketch)
